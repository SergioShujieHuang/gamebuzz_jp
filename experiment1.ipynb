# Task
Perform a logistic regression analysis using the data in `df` to determine if the proportion of time spent reading the original text affects the likelihood of noticing errors.

First, preprocess the data from "./drive/MyDrive/file_backup/Master/研究/実験１/Book1.xlsx":
1. Convert time columns (e.g., '原文読む時間', '要約読む時間') from 'XXmXXsXX' format to total seconds.
2. Calculate the 'proportion of time spent reading the original text' relative to the total reading time (original + summary).
3. Convert the '誤りを気づいた' column from 'T/F' to binary (1/0).

Next, execute a logistic regression using the `statsmodels` library with '誤りを気づいた' as the dependent variable and the 'time proportion' as the independent variable.

Finally, output the results including the Wald Z-test statistics, confidence intervals, and the effect size (odds ratio), accompanied by an interpretation of the statistical significance in Japanese.

## data_preprocessing

### Subtask:
Convert time columns to seconds, calculate the original reading time proportion, and convert the target variable to binary.


```python
import re

def parse_time_to_seconds(time_str):
    if pd.isna(time_str) or not isinstance(time_str, str):
        return 0.0
    # Matches formats like '2m22s9' or '0m28s99'
    match = re.match(r'(\d+)m(\d+)s(\d+)', time_str)
    if match:
        minutes = int(match.group(1))
        seconds = int(match.group(2))
        centiseconds = int(match.group(3))
        # Convert everything to seconds
        return minutes * 60 + seconds + centiseconds / 100.0
    return 0.0

# 1 & 2. Convert time columns to numerical seconds
df['原文読む時間_sec'] = df['原文読む時間'].apply(parse_time_to_seconds)
df['要約読む時間_sec'] = df['要約読む時間'].apply(parse_time_to_seconds)

# 3. Calculate original reading time proportion
df['original_reading_proportion'] = df['原文読む時間_sec'] / (df['原文読む時間_sec'] + df['要約読む時間_sec'])

# 4. Map '誤りを気づいた' to binary (T=1, F=0)
df['誤りを気づいた_binary'] = df['誤りを気づいた'].map({'T': 1, 'F': 0}).astype(float)

# 5. Inspect the result
print("Preprocessed DataFrame:")
display(df[['原文読む時間', '原文読む時間_sec', '要約読む時間', '要約読む時間_sec', 'original_reading_proportion', '誤りを気づいた_binary']].head())
```

## logistic_regression_analysis

### Subtask:
Perform a logistic regression using statsmodels to analyze the impact of original reading time proportion on error detection.


```python
import statsmodels.api as sm
import numpy as np

# 1. Define independent (X) and dependent (y) variables
# Adding a constant for the intercept term
X = sm.add_constant(df['original_reading_proportion'])
y = df['誤りを気づいた_binary']

# 2. Fit the logistic regression model
model = sm.Logit(y, X)
results = model.fit()

# 3. Calculate Odds Ratios and Confidence Intervals
odds_ratios = np.exp(results.params)
conf_int = results.conf_int()
odds_conf_int = np.exp(conf_int)

# 4. Consolidate results for interpretation
odds_df = pd.DataFrame({
    'Odds Ratio': odds_ratios,
    'Lower CI (95%)': odds_conf_int[0],
    'Upper CI (95%)': odds_conf_int[1]
})

# 5. Output the results
print(results.summary())
print("\n--- Odds Ratios and Confidence Intervals ---")
display(odds_df)

# Japanese interpretation based on p-value
p_value = results.pvalues['original_reading_proportion']
if p_value < 0.05:
    print(f"\n統計的有意性: 有意水準5%で、原文読む時間の割合が誤りを気づく確率に有意な影響を与えていると言えます (p = {p_value:.4f})。")
else:
    print(f"\n統計的有意性: 有意水準5%で、原文読む時間の割合が誤りを気づく確率に有意な影響を与えているとは言えません (p = {p_value:.4f})。")
```

## calculate_scores_and_covariate

### Subtask:
Calculate accuracy scores for the three question groups and verify the covariate for the analysis.


```python
import re

def extract_score(text):
    if pd.isna(text):
        return 0.0
    if isinstance(text, (int, float)):
        return float(text)
    # Matches patterns like '10.00 / 10' or just '10.00'
    match = re.search(r'(\d+\.?\d*)\s*/\s*\d+', str(text))
    if match:
        return float(match.group(1))
    match_simple = re.search(r'(\d+\.?\d*)', str(text))
    if match_simple:
        return float(match_simple.group(1))
    return 0.0

# Identify columns containing '[Score]'
score_cols = [col for col in df.columns if '[Score]' in col]

# Apply extraction to all score columns
for col in score_cols:
    df[f'{col}_numeric'] = df[col].apply(extract_score)

# Map specific questions based on their content keywords found in the column list
# 屈折角 related: '虹が「色の帯」として見えるのは...屈折角... [Score]'
# 二重虹 related: '「二重虹（外側の虹）」が形成される際... [Score]'
refraction_col = [c for c in score_cols if '屈折角' in c][0] + '_numeric'
double_rainbow_col = [c for c in score_cols if '二重虹' in c and '反射' in c][0] + '_numeric'

specific_2_cols = [refraction_col, double_rainbow_col]
remaining_8_cols = [f'{c}_numeric' for c in score_cols if f'{c}_numeric' not in specific_2_cols][:8]
final_10_cols = specific_2_cols + remaining_8_cols

# Calculate scores (assuming max score per question is 10, normalize to 0-1 range)
df['total_score'] = df[final_10_cols].mean(axis=1) / 10.0
df['specific_2_score'] = df[specific_2_cols].mean(axis=1) / 10.0
df['remaining_8_score'] = df[remaining_8_cols].mean(axis=1) / 10.0

print("Scores calculated successfully.")
print(f"Questions used: {len(final_10_cols)}")
display(df[['total_score', 'specific_2_score', 'remaining_8_score', 'original_reading_proportion']].head())
```

## mixed_design_ancova_analysis

### Subtask:
Perform a mixed-design ANCOVA or separate ANCOVAs to analyze the effect of noticing errors on test scores, using the reading time proportion as a covariate.


```python
try:
    import pingouin as pg
except ImportError:
    !pip install pingouin
    import pingouin as pg

# 1. ANCOVA for Specific 2 Score
print("--- ANCOVA for Specific 2 Score ---")
ancova_specific = pg.ancova(data=df, dv='specific_2_score', covar='original_reading_proportion', between='誤りを気づいた')
display(ancova_specific)

# 2. ANCOVA for Remaining 8 Score
print("\n--- ANCOVA for Remaining 8 Score ---")
ancova_remaining = pg.ancova(data=df, dv='remaining_8_score', covar='original_reading_proportion', between='誤りを気づいた')
display(ancova_remaining)

# 3. ANCOVA for Total Score
print("\n--- ANCOVA for Total Score ---")
ancova_total = pg.ancova(data=df, dv='total_score', covar='original_reading_proportion', between='誤りを気づいた')
display(ancova_total)

# Descriptive stats for context
print("\n--- Descriptive Statistics by Group ---")
desc_stats = df.groupby('誤りを気づいた')[['specific_2_score', 'remaining_8_score', 'total_score']].agg(['mean', 'std'])
display(desc_stats)
```

### ANCOVA Analysis Summary
